In this exercise, we will use K-means clustering to do image compression

```python
#Importing dependencies
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import imread
from scipy.misc import imsave

%matplotlib inline
```

#### Select 'K' random points as centroids

```python
def generateCentroids(X,K):
	"""
	Randomly permutes X and returns the first K rows as centroids
	Input:
	X - Training set
	K - Number of clusters (centroids)
	Output:
	centroid - List of tuples of form (x,y) where x,y are co-ordinates of centroids
	"""

	permuted_X = np.random.permutation(X)
	return [tuple([permuted_X[i][j] for j in range(X.shape[1])]) for i in range(K)]
```

#### Find closest centroid for every observation

```python
def findClosestCentroids(X,initial_centroids):
	"""
	Returns array c that stores the index of closest centroid for every observation in training set
	Input:
	X - Training set
	initial_centroids - List of tuples indicating position of centroids
	Every tuple in list is of form (x,y) where x,y are co-ordinates of centroid
	Output:
	c - array that stores the index of closest centroid for every observation in training set
	"""

	#The c array corressponding to the c array in programming exercise
	c = []

	for curr_x in X:
		#Initializing new position in c array with -1
		c.append(-1)
		#temp_dist stores the distance of current example in X with the closest centroid
		#Initialized to infinity as initially, this distance is not known
		temp_dist = np.inf
		for cent_index,centroid in zip(range(len(initial_centroids)), initial_centroids):
			curr_dist = np.linalg.norm(np.array(curr_x) - np.array(centroid))**2
			if curr_dist<temp_dist:
				temp_dist = curr_dist
				c[-1]=cent_index
	return c
```

#### Computing new position of centroids

```python
def computeCentroids(X, c, K):
	"""
	Returns list containing new position of centroids
	Input - 
	X - Training set
	c - array that stores the index of closest centroid for every observation in training set (X)
	K - number of clusters (centroids)
	Output - 
	centroids - list of tuples that stores new position of centroids
	tuples are of form (x,y) where x,y are co-ordinates of centroids
	"""

	X = np.array(X)
	c = np.array(c)
	centroids = []
	for i in range(K):
		samples = X[np.where(c==i)]
		curr_centroid_mean = np.mean(samples,axis=0)
		centroids.append(tuple([curr_centroid_mean[i] for i in range(X.shape[1])]))
	return centroids
```

#### Method to run k-means on given data for given number of iterations


```python
def runkmeans(X, initial_centroids, max_iters, plot_progress=True):
	"""
	Runs the kmeans algorithms for max_iters times and returns the final cluster assignments
	and centroid positions optionally plotting the progress.
	Input:
	X - Training set
	initial_centroids - Initial position of centroids
	max_iters - Number of iterations for which to run kmeans algorithms
	plot_progress - Plots the progress of algorithm when set to True
	Output:
	(centroids, c) - Tuple containing centroids and c arrays where 
	centroids - List of tuples containing position of centroids after running kmeans
	tuples are of form (x,y) where x,y are co-ordinates of centroids
	c - Array indicating cluster assignment after running kmeans
	"""

	#To enable interactive plotting
	if plot_progress:
		plt.ion()   
		plt.figure()

	centroids = np.array(initial_centroids)
	c = np.array(findClosestCentroids(X,initial_centroids))
	#plot_centroids = [[centroids]]

	for i in range(max_iters):
		print("Starting iteration " + str(i+1))
		if plot_progress:
			plt.clf()
			colors = ['r','g','b']
			for curr_cent,color in zip(range(len(centroids)), colors):
				plt.scatter(X[np.where(c==curr_cent)][:,0], X[np.where(c==curr_cent)][:,1], facecolors='none', edgecolors=color, s=30)
				plt.plot(centroids[curr_cent][0], centroids[curr_cent][1], marker='x', color='black', markersize=10, markeredgewidth=2)
				#cents = [plot_centroids[i] for i in range(curr_cent,len(plot_centroids), 3)]
				#print(cents)
				#plt.plot(cents, marker='x', color='black')
			plt.title('Iteration number ' + str(i+1))	

		c = np.array(findClosestCentroids(X,centroids))
		centroids = np.array(computeCentroids(X, c, len(initial_centroids)))
		#plot_centroids.append([centroids])
		#print(plot_centroids.shape)

		if plot_progress:
			enter = input("Press enter to continue")
		else:
			print("Done iteration " + str(i+1))
	return (centroids, c)
```

##### Export this notebook as python file and run it in terminal/cmd to see visualization of cluster formation with moving centroids

```python
#Initializing data
data = np.array(loadmat('ex7data2.mat')['X'])
#Setting K (number of clusters)
K = 3

initial_centroids = generateCentroids(data, K)
#Running k-means for 10 iterations
max_iters = 10

#Finding closest centroids for every datapoint
idx = findClosestCentroids(data, initial_centroids)
#Finding initial cluster assignment
centroids = computeCentroids(data, idx, K)

#Running k-means
centroids, c = runkmeans(data, initial_centroids, max_iters)
```

#### Image Compression using k-means

```python
#Loading image
img = imread('bird_small.png')
ori_x = img.shape[0]
ori_y = img.shape[1]
img = img.reshape(-1,3)
#Normalizing image pixel values
img = img/255
```

```python
#Forming 16 clusters
K=16
#Running k-means for 10 iterations
max_iters = 10
#Generating initial centroids
init_centroids = generateCentroids(img,K)
#Initial cluster assignment
c = findClosestCentroids(img,init_centroids)
#Starting kmeans
final_centroids, final_c = runkmeans(img, init_centroids, max_iters, plot_progress=False)
```

```python
#Saving compressed image
compressed_img = np.zeros([ori_x*ori_y, 3])

for i in range(K):
	compressed_img[final_c==i] = final_centroids[i]

compressed_img = compressed_img.reshape(ori_x, ori_y, 3)
imsave('compressed_img.png', compressed_img)
```

```python
#Visualization - showing differences between orignal and compressed image
plt.subplot(121)
plt.imshow(img.reshape(ori_x, ori_y, 3))
plt.title("Original Image")

plt.subplot(122)
plt.imshow(compressed_img)
plt.title("Compressed Image")

plt.tight_layout()
```

#### PCA

```python
def project_data(X, U, K):
	ureduce = U[:,:K]
	return np.dot(X,ureduce)
```

```python
def recover_data(Z,U,K):
	return np.dot(Z,U[:,:K].T)
```

```python
#Initializing data
data = np.array(loadmat('ex7data1.mat')['X'])
```

```python
means = np.mean(data, axis=0)
stds = np.std(data, axis=0)
normalized_data = (data - means)/stds
co_var = (np.dot(normalized_data.T, normalized_data))/normalized_data.shape[0]
```

```python
u,s,v = np.linalg.svd(co_var)
```

```python
# Visualization
plt.scatter(data[:,0], data[:,1], facecolors='none', edgecolor='blue')

plt.plot([means[0], means[0] + 1.5*s[0]*u[0,0]], 
         [means[1], means[1] + 1.5*s[0]*u[0,1]], color='black', markersize=10)
plt.plot([means[0], means[0] + 1.5*s[1]*u[1,0]], 
         [means[1], means[1] + 1.5*s[1]*u[1,1]], color='black', markersize=10)
```

```python
k=1
z = project_data(normalized_data, u, k)
print(z)
```

```python
recovered = recover_data(z,u,k)
print(recovered)
```
