## Anomaly Detection
---
In this exercise, we will perform Anomaly Detection to detect failing servers on a network.

```python
#Importing dependencies
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
%matplotlib inline
```

```python
def estimateGaussian(X):
	mu = np.mean(X, axis=0, keepdims=True)
	sigma2 = np.sum(np.square(X - mu), axis=0, keepdims=True)/X.shape[0]
	return (mu, sigma2)
```

```python
def gaussian(X, mu, sigma2):
	#Using the original model and not the multivariate gaussian model
	first = (1/np.sqrt(2*np.pi * sigma2))
	second_num_one = np.square(X[:,0] - mu[0][0])
	second_num_two = np.square(X[:,1] - mu[0][1]) 
	second_den = 2*sigma2
	second_one = np.exp((-second_num_one)/second_den[0][0])
	second_two = np.exp((-second_num_two)/second_den[0][1])
	z_one = first[0][0]*second_one
	z_two = first[0][1]*second_two
	return z_one*z_two
```

```python
def gaussian_vectorized(X, mu, sigma2):
	#Using the original model and not the multivariate gaussian model
	first = (1/np.sqrt(2*np.pi * sigma2))
	second = np.exp( (-np.square(X - mu)) / (2*sigma2))
	Z = first*second
	return Z.prod(axis=1)
```

```python
def visualize(X, mu, sigma2):

	xx, yy = np.meshgrid( np.arange(0, 30, 0.5), np.arange(0,30,0.5) )
	plot_X = np.c_[xx.ravel(), yy.ravel()]

	Z = gaussian_vectorized(plot_X, mu, sigma2)
	Z = Z.reshape(xx.shape)

	plt.scatter(X[:,0], X[:,1], marker='x', linewidth=0.5)
	plt.xlabel('Latency(ms)')
	plt.ylabel('Throughput(mb/s)')
	plt.contour(xx,yy,Z,[10**exp for exp in range(-20,0,3)])
	plt.axis([0,30,0,30])
```

```python
def F1_score(yval, pval):

	tp = np.sum(np.logical_and(pval==1, yval==1)).astype(float)
	fp = np.sum(np.logical_and(pval==1, yval==0)).astype(float)
	fn = np.sum(np.logical_and(pval==0, yval==1)).astype(float)

	p = tp/(tp+fp) if (tp+fp)>0 else 0
	r = tp/(tp+fn) if (tp+fn)>0 else 0

	return ((2*p*r)/(p+r)) if (p+r)>0 else 0
```

```python
def selectThreshold(yval, pval):
	F1 = 0
	epsilon = 0

	step = (pval.max() - pval.min())/1000
	for curr_epsilon in np.arange(pval.min(), pval.max(), step):

		binary_pval = pval < curr_epsilon
		curr_f1 = F1_score(yval, binary_pval)
		if curr_f1>F1:
			F1 = curr_f1
			epsilon = curr_epsilon

	return (epsilon, F1)
```

```python
def disp_anomalies(X, mu, sigma2, epsilon):
    z = gaussian_vectorized(X, mu, sigma2)
    plot_X = X[z<epsilon]
    plt.scatter(plot_X[:,0], plot_X[:,1], facecolor='none', edgecolor='r', s=100, linewidth=2)
    plt.axis([0,30,0,30])
```

#### Part 1

```python
#Initializing data
data = loadmat('ex8data1.mat')

X = np.array(data['X'])
Xval = np.array(data['Xval'])
yval = np.array(data['yval'])
```

```python
mu, sigma2 = estimateGaussian(X)
print("Mu value:" + str(mu))
print("sigm2 value:" + str(sigma2))
visualize(X, mu, sigma2)
```

```python
#Selecting thresholds
pval = gaussian_vectorized(Xval, mu, sigma2)
yval = np.array(yval, np.float64)
pval = pval.reshape(yval.shape)


epsilon, F1 = selectThreshold(yval, pval)
print(epsilon)
print(F1)

#Displaying anomalies
visualize(X, mu, sigma2)
disp_anomalies(X, mu, sigma2, epsilon)
```
